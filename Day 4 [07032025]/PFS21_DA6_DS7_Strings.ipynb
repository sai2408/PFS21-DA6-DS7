#Strings
    A string is a sequence of characters of immutable data type
    String literals in python are surrounded by either single quotation marks, or double quotation marks.

    
    Operations on Strings:
        Creation
        Accessing

```python
S = "HELLO"
print(S)
s1 = 'world'
print(s1)
```

#Creation
    Creation of an Empty String
    Creation of String with Single Element
    Creation of String with multiple Elements
    Creation of String with user Inputs

```python
s1 = ""
s2 = ''
print(s1)
print(s2)
s3 = "a"
s4 = 'a'
print(s3)
print(s4)
s5 = "abc"
s6 = 'abc'
print(s5)
print(s6)
```

#Methods in Strings
    1. endswith()
    2. startswith()
    3. replace()
    4. isdigit()
    5. isalpha()
    6. isalnum()
    7. istitle()
    8. isupper()
    9. islower()
    10. upper()
    11. lower()
    12. swapcase()
    13. capitalize()
    14. find()
    15. rfind()
    16. count()


```python
#endswith
'''
print(x.endswith("o"))
'''
#startswith
'''
x = "hello"
print(x.startswith("x"))
print(x.startswith("h"))
print(x.startswith("he"))
print(x.startswith("hello"))
'''
#replace
'''
s1 = "python"
s2 = "hello"
print(s1.replace("p","c"))
print(s1.replace("x","y"))
print(s2.replace("l","t"))
'''
#isdigit
'''
s1 = "10"
s2 = "hello"
print(s1.isdigit())
print(s2.isdigit())
'''
#isalpha
'''
s1 = "10"
s2 = "hello"
s3 = "hello123"
print(s1.isalpha())
print(s2.isalpha())
print(s3.isalpha())
'''
#isalnum
'''
s1 = "10"
s2 = "hello"
s3 = "hello123"
s4 = "hello$world"
print(s1.isalnum())
print(s2.isalnum())
print(s3.isalnum())
print(s4.isalnum())
'''
#istitle
'''
s1 = "hello world"
s2 = "Hello world"
s3 = "Hello World"
print(s1.istitle())
print(s2.istitle())
print(s3.istitle())
'''
#isupper
'''
s1 = "GHello"
s2 = "HELLO"
print(s1.isupper())
print(s2.isupper())
'''
#islower
'''
s1 = "hello"
s2 = "HELLO"
print(s1.islower())
print(s2.islower())
'''
#upper
'''
s1 = "Hello"
print(s1.upper())
'''
#lower
'''
s1 = "Hello"
print(s1.lower())
'''
#swapcase
'''
s1 = "aAbBcC"
print(s1.swapcase())
'''
#capitalize
'''
s1 = "i am sai"
print(s1.capitalize())
'''
#find
'''
s1 = "python cython"
print(s1.find("t"))
print(s1.find("hon"))
'''
#rfind
'''
s1 = "python cythonn"
print(s1.rfind("t"))
print(s1.rfind("hon"))
'''
#count
'''
s1 = "python cython"
print(s1.count(" "))
print(s1.count("p"))
print(s1.count("thon"))
print(s1.count("y"))
'''
```
