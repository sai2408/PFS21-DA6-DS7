#BITWISE OPERATOR

image.png

```python
print(5 & 2)
print(5 & 1)
print(5 & 3)
print(8 & 9)
print(5 | 2)
print(5 ^ 1)
print(5 ^ 7)
print(5 >> 3)
print(2<<1)
print(2<<2)
print(2<<3)
```

#LOGICAL OPERATORS

image.png

```python
a = 10
b = 20
print((a>0) and (a>b))
print((a>b) and (a>0))

print((a>0) or (a>b))
print((a>b) or (a>0))

print(not((a>b) and (a>0)))
print(not((a>b) or (a>0)))

print(not(a>0) and (b>a))
```
