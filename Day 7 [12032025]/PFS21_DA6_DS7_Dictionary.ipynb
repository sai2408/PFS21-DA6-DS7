#Dictionary
    1. A dictionary is aÂ collection which is unordered, changeable and indexed.
    2. In Python dictionaries are written with curly brackets, and they have keys and values.
    3. This data structure is combination of both mutable and immutable data types.
    4. Keys are immutable
    5. Values are mutable
    
    Operations on Dictionaries:
        1. Creation
        2. Accessing
        3. Insertion
        4. Deletion
        5. Updation

#Creation
    1. Creation of an Empty Dictionary
    2. Creation of dictionary with Single Element
    3. Creation of dictionary with multiple Elements
    4. Creation of dictionary with user Inputs

```python
'''
d = {}
print(d)
print(type(d))
'''
'''
d = {"hi":100.56}
print(d)
print(type(d))
'''
'''
d = {10:100,40:"hello","hi":"abcd","10":656.677}
print(d)
print(type(d))
'''
```

#Accessing
    Indexing
        Using Keys
    Looping
        For Loop with range
        For Loop without range
    Methods
        get()
        copy()
        keys()
        items()
        values()

```python
'''
d1 = {10:100,20:200,30:300,40:400}
print(d1[20])
print(d1[21])
'''
d1 = {1:100,2:200,3:300,4:400,5:500}
'''
print(d1[4])
print(d1.get(4))
print(d1.get(10))
'''
'''
d2 = d1.copy()
print(d2)
'''
'''
print(d1.keys())
print(d1.values())
print(d1.items())
'''
```

#Insertion
    Indexing
        Positive Indexing
    Methods
        update()

```python
'''
d1 = {10:100,20:200,30:300,40:400}
print(d1)
d1[20] = 250
print(d1)
d1[30] = 30
print(d1)
'''
'''
d1 = {10:100,20:200,30:300}
d2 = {40:400,50:500,60:600}
d1.update(d2)
print(d1)
'''
```

#Deletion
    Methods
        pop(key)
        popitem()
        clear()
        del dictionary_name[key]
        del(d)

```python
d1 = {10:100,20:200,30:300}
'''
print(d1)
d1.pop(20)
print(d1)
'''
'''
print(d1)
d1.popitem()
print(d1)
'''
'''
print(d1)
d1.clear()
print(d1)
'''
'''
print(d1)
del d1[20]
print(d1)
'''
'''
print(d1)
del d1
print(d1)
'''
```

#Nested Dictionary
    1. Dictionary inside another dictionary is called nested dictionary.
    2. Nested value can be of any data type.


```python
d1 = {10:[10,20,30],20:["hello","world"],30:[True,False]}
print(d1)
print(d1[20])
print(d1[10][1])
```

```python
d1 = {10:100,30:200,20:300}
print(10 in d1)
print(100 in d1)
print(sorted(d1))
```
