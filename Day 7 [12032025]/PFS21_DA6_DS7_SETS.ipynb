#Sets
    1. A Set is an unordered collection data type that is iterable, mutable and has no duplicate elements.
    2. Set is majorly used to perform mathematical set operations like union,difference,intersection etc,..
    
    Operations on sets:
        1. Creation
        2. Accessing
        3. Insertion
        4. Deletion
        5. Updation

#Creation
    Creation of an Empty Set
    Creation of Set with Single Element
    Creation of Set with multiple Elements
    Creation of Set with user Inputs

```python
'''
s1=  set()
print(s1)
print(type(s1))
'''
'''
s1 = {"hello"}
print(s1)
s2 = set([10])
print(s2)
'''
s1 = {10,20,30,40,5.67,"hello"}
print(s1)
s2 = set([10,20,30,40,5.67,"hello"])
print(s2)
s3 = {10,20,30,40,20}
print(s3)
s4 = set([10,20,30,40,20])
print(s4)
```
