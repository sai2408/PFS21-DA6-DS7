#Tuples
    Tuple is ordered collection of items.
    It could be of any python data type
    Its an immutable data structure

    Operations on Tuple:
        Creation
        Accessing

#Creation
    Creation of an Empty Tuple
    Creation of Tuple with Single Element
    Creation of Tuple with multiple Elements
    Creation of Tuple with user Inputs [ Lets Discuss Later ]

```python
'''
x = ()
print(x)
print(type(x))
'''
'''
x = ("abc",)
print(x)
print(type(x))
'''
'''
x = (10,"56.78",56.78,"hello")
print(x)
print(type(x))
'''
```

#Accessing
    Indexing
        Positive Indexing
        Negetive Indexing
    Looping
        For Loop with range
        For Loop without range
    Methods
        index()
        count()
    Slicing
        Positive Slicing
        Negetive Slicing

```python
x = (10,20,30,40,50,60,60,70,80,90)
print(x)
print(type(x))

#Indexing[+ve,-ve]
'''
print(x[1])
print(x[5])
print(x[-1])
print(x[-4])
'''

'''
x = (10,20,30,40)
print(x.index(30))
print(x)
'''

'''
y = (10,20,30,20,40,20)
print(y.count(20))
print(y)
'''
X = (10,20,30,40,50,60,70,80,90)
print(x[::])
print(x[3::])
print(x[3:7])
print(x[3:9:2])
```

#Built-in functions in Tuple
    len()
    max()
    min()
    sum()
    sorted()

```python
x = (5,3,7,23,0)
print(len(x))
print(max(x))
print(min(x))
print(sum(x))
print(sorted(x))
```

#External Operations on Tuple
    concatenation
    repetetion
    membership

```python
x1 = (1,2,3)
x2 = (4,5,6)
print(x1)
print(x2)
print(x1+x2)
print(x1)
print(x2)
print(x1*5)
print(1 in x1)
print(5 in x1)
```

#Nested Tuple
    Tuple inside another tuple is called as tuple

```python
z = (1,2,(11,12,13),6)
print(z[3])
print(z[2][1])
```
