```python
content = soup.find('div', attrs={'class' : 'main'})
#print(content)

text=content.text
#print(text)
r1 = text.split("\n")
#print(r1)

r2 = []
for i in r1:
  if i != '' and i != '\r':
    r2.append(i)
#print(r2)
r3 = []
for i in r2:
  k = i.strip()
  r3.append(k)
#print(r3)
r4 = []
for i in r3:
  if i != '':
    r4.append(i)
#print(r4)
names = []
prices = []
for i in range(len(r4)):
  if i%2 == 0:
    names.append(r4[i])
  else:
    prices.append(r4[i])

print(names)
print(prices)
print(len(names))
print(len(prices))
```

#Create a DataFrame and save it in CSV file

```python
df = pd.DataFrame({'Product_Name' : names,'MRP' : prices})
```
