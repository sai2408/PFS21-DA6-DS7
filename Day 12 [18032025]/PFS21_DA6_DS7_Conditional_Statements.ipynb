#**Conditional Statements**
    Conditional statements are also known as decision-making statements.
    Conditional statements control the flow of a program by executing different blocks of code based on whether a
    condition evaluates to True or False.
#**Types of Conditional Statements in Python**
    IF Statement
    IF - ELSE Statement
    IF - ELIF - ELSE Statement
    Nested Conditional Statements

#**IF STATEMENT**

```python
print("Line 1")
print("Line 2")
if 100>20:
  print("IF 1")
  print("IF 2")
  print("IF 3")
print("Line n-1")
print("Line n")
```

#**IF ELIF ELSE STATEMENT**


```python
print("Line 1")
print("Line 2")
if 10>20:
  print("IF 1")
  print("IF 2")
  print("IF 3")
elif 10==100:
  print("ELIF 1")
  print("ELIF 2")
else:
  print("ELSE 1")
  print("ELSe 2")
print("Line n-1")
print("Line n")
```

#**CONDITIONAL EXPRESSION [ Python's Ternary Operator ]**

#**Syntax**
    <expression 1> if <Condition> else <expression 2>

```python
n = 12
if n%2==0:
  print("Even")
else:
  print("Odd")

x = "Even" if n%2==0 else "Odd"
print(x)
print("Even" if n%2==0 else "Odd")
```
